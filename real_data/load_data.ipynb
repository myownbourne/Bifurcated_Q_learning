```python
import numpy as np
import pandas as pd

mimictable = pd.read_csv('C:\\Users\\chen kai\\Desktop\\DTR_code\\mimic\\mimic_sepsis-main\\MIMICtable_new.csv')
unique_ids = mimictable['icustayid'].unique() 
sol_ids = []
print(len(unique_ids))

def check_vector(vector):
    set1 = {1, 2, 3, 4, 5, 6}
    set2 = {7, 8, 9, 10, 11, 12}
    return bool(set(vector) & set1) and bool(set(vector) & set2)


for id in unique_ids:
	filtered_rows = mimictable[mimictable['icustayid'] == id]
	a = check_vector(filtered_rows['bloc'].values)
	if a:
		sol_ids.append(id)

data1 = mimictable[(mimictable['icustayid'].isin(sol_ids)) & (mimictable['bloc'].isin([1,2,3,4,5,6]))]
data2 = mimictable[(mimictable['icustayid'].isin(sol_ids)) & (mimictable['bloc'].isin([7,8,9,10,11,12]))]
id1 = data1['icustayid'].unique()
```

```python
data1_g = data1.groupby('icustayid')
mean_names = data1.columns[~data1.columns.isin(['icustayid', 'input_total', 'output_total'])].tolist()
avg_values = data1_g[mean_names].mean()
last_input_total = data1_g.apply(lambda g: g.loc[g['bloc'].idxmax(), 'input_total'])
last_output_total = data1_g.apply(lambda g: g.loc[g['bloc'].idxmax(), 'output_total'])
```

```python
df1 = pd.concat([avg_values, last_input_total, last_output_total], axis=1)
df1.columns = mean_names + ['input_total', 'output_total']
df1['mechvent'] = df1['mechvent'].apply(lambda x: 1 if x > 0 else x)
df1['re_admission'] = df1['re_admission'].apply(lambda x: 1 if x > 0 else x)
df1.reset_index(inplace=True)
```

```python
data2_g = data2.groupby('icustayid')
mean_names2 = data2.columns[~data2.columns.isin(['icustayid', 'input_total', 'output_total'])].tolist()
avg_values2 = data2_g[mean_names].mean()
last_input_total2 = data2_g.apply(lambda g: g.loc[g['bloc'].idxmax(), 'input_total'])
last_output_total2 = data2_g.apply(lambda g: g.loc[g['bloc'].idxmax(), 'output_total'])
```

```python
df2 = pd.concat([avg_values2, last_input_total2, last_output_total2], axis=1)
df2.columns = mean_names2 + ['input_total', 'output_total']
df2['mechvent'] = df2['mechvent'].apply(lambda x: 1 if x > 0 else x)
df2['re_admission'] = df2['re_admission'].apply(lambda x: 1 if x > 0 else x)
df2.reset_index(inplace=True)
```

```python
y_df = df2[['icustayid','died_in_hosp','mortality_90d']]
```

```python
df1.to_csv('S1_24_24_comp2.csv', index=False) 
df2.to_csv('S2_24_24_comp2.csv', index=False) 
y_df.to_csv('y_24_24_comp2.csv', index=False) 
```
